```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#loading boston dataset and printing

boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target, columns=["target"])
X['MEDV'] = boston.target
X.head(50)
```

```python
col1 = pd.DataFrame(X, columns=["CRIM"])
hist1 = col1.hist(bins=40)
```

```python
col2 = pd.DataFrame(X, columns=["RM"])
hist2 = col2.hist(bins=20)
```

```python
#plots of the given data to make the first assumptions of its behaviour

features = ['RM', 'CRIM']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = X[col]
    plt.scatter(x, y, marker='o')
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

The first picture - plot of RM - shows that the target is positively and linearly dependant on the column (average number of rooms per dwelling). But the second one, which represents per capita crime rate by town, fits negatively to the target and is more like an exponential function. Now let's see how the linear regressions on them look like.

```python
#linear regression on column 'RM'

x = pd.DataFrame(np.c_[X['RM']], columns = ['RM'])

rm = LinearRegression()
rm.fit(x, y)
y_pred = rm.predict(x)

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, y_pred, color='r')
plt.show()
```

```python
#linear regression on column 'CRIM'

x = pd.DataFrame(np.c_[X['CRIM']], columns = ['CRIM'])

rm = LinearRegression()
rm.fit(x, y)
y_pred = rm.predict(x)

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, y_pred, color='r')
plt.show()
```

As we assumed before, the first linear regression fit its plot strongly. But the next one is not the right fit for these data. Actually, the target is evidently dependant on the CRIM data, but in another way - we can put a descending exponential function to fit the points, rather than the linear one.
